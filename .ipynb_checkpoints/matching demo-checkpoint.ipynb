```python
import sys
sys.path.append('yolov7_bbox')
from face_alignment import *
import yaml
import torch
from yolov7_bbox.models2.yolo import Model
from yolov7_bbox.bbox_detection import detect as bbox_detect

import sys
sys.path.append('yolov7_bbox_landmarks')

from yolov7_bbox_landmarks.models.experimental import attempt_load
from yolov7_bbox_landmarks.kpts_detection import detect as kpt_detect

import iresnet
import torch
from torch import nn
import shutil

import cv2
import numpy as np
from PIL import Image
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt

@torch.no_grad()
def inference(net, img):
    if img is None:
        img = np.random.randint(0, 255, size=(112, 112, 3), dtype=np.uint8)
    else:
        img = cv2.resize(img, (112, 112))

    img = np.transpose(img, (2, 0, 1))
    img = torch.from_numpy(img.copy()).unsqueeze(0).float()
    img.div_(255).sub_(0.5).div_(0.5)
    net.eval()
    with torch.no_grad():
        feat = net(img.to('cuda'))
    return feat

def save_yaml(data, file_path):
    with open(file_path, 'w') as file:
        yaml.dump(data, file)

def load_yaml(file_path):
    with open(file_path, 'r') as file:
        data = yaml.safe_load(file)
    return data

def pet_face_detector(img_path):
    img = cv2.imread(img_path)
    imgs = []
    _, bbox_list, src_pts_list, conf = kpt_detect(yolo_model1, img_path, device)
    _, bbox_list2, conf2 = bbox_detect(yolo_model2, img_path, device)
    if src_pts_list and bbox_list:
        for i, (bbox, s) in enumerate(zip(bbox_list, src_pts_list)):
            src_pts = adjust_src_pts(np.array(s).reshape(3, 2))
            face_img = preprocess(img, bbox, src_pts, image_size="112, 112", margin=0)
            imgs.append(face_img)
    elif bbox_list2:
        for i, bbox in enumerate(bbox_list2):
            crop_img = img[int(bbox[1]):int(bbox[3]), int(bbox[0]):int(bbox[2])]
            imgs.append(crop_img)
    return imgs
```

```python
file_path = 'yolov7_bbox/yolo_model2_yaml.yaml'
yolo_model2_yaml = load_yaml(file_path)

weights = 'yolov7_bbox/weights/yolov7x-pet-face.pth'
state_dict = torch.load(weights)
state_dict = {key[6:]: value for key, value in state_dict.items()}

device = 'cuda' if torch.cuda.is_available() else 'cpu'
yolo_model2 = Model(yolo_model2_yaml).to(device)
yolo_model2.load_state_dict(state_dict)
yolo_model2.autoshape()
print(device)
```

```python
model_path = 'best_backbone_2023_12_19.pth'
device = 'cuda'
backbone = iresnet.iresnet100(num_classes=512)
backbone.load_state_dict(torch.load(model_path, map_location=torch.device(device)))
backbone.to(device)
```

```python
src = f'registrated_pet/imgs'
dst = f'registrated_pet/embs'

backbone.to(device)
for f in os.listdir(src):
    os.makedirs(f'{dst}/{f}', exist_ok=True)
    for img_name in os.listdir(f'{src}/{f}'):
        emb_file = f'{dst}/{f}/{img_name}.npy'
        if os.path.exists(emb_file):
            continue
        img = cv2.imread(f"{src}/{f}/{img_name}")[..., ::-1]
        emb = inference(backbone, img).detach().cpu().numpy()
        with open(emb_file, 'wb') as ef:
            np.save(ef, emb)
```

```python
## load embs
import pandas as pd

df = pd.read_csv('registrated_pet/cat_or_dog.csv')
cls_dict = {k: v for k, v in df[['path', 'cls']].values}

src = f'registrated_pet/embs'
labels = []
embs = []
gt_path = []
cls_list = []

for f in os.listdir(src):
    for emb_name in os.listdir(f'{src}/{f}'):
        emb_file = f'{src}/{f}/{emb_name}'
        with open(emb_file, 'rb') as ef:
            emb = np.load(ef)
        embs.append(emb)
        labels.append(f)
        gt_path.append(f"registrated_pet/imgs/{f}/{emb_name.replace('.npy', '')}")
        cls_list.append(cls_dict[emb_name.replace('.npy', '')])

embs = torch.from_numpy(np.array(embs)).squeeze(1)
labels = np.array(labels)
gt_path = np.array(gt_path)
cls_list = np.array(cls_list)
```

```python
src = f'upload_pet_img/anchor_the_shepherd'
cls = 'dog'

cos = nn.CosineSimilarity(eps=1e-6).to(device)

test_embs = []
test_imgs = []
for img_name in os.listdir(src):
    img_path = f'{src}/{img_name}'
    imgs = pet_face_detector(img_path)
    for img in imgs:
        test_imgs.append(img)
        emb = inference(backbone, img[..., ::-1]).detach().cpu().numpy()
        test_embs.append(emb)

test_embs = torch.from_numpy(np.array(test_embs))
```

```python
test_embs.shape
```

```python
y_pred = []
y_cos = []
pred_path = []

mask = cls == cls_list

for test_emb in test_embs:
    res = cos(test_emb, embs[mask])
    y_cos += list(res.numpy())
    y_pred += list(labels[mask])
    pred_path += list(gt_path[mask])

inds = np.argsort(y_cos)[::-1]

n = 5

top_n_cos = []
top_n_pred = []
top_n_pred_path = []

for ind in inds:
    if y_pred[ind] not in top_n_pred:
        top_n_pred.append(y_pred[ind])
        top_n_cos.append(y_cos[ind])
        top_n_pred_path.append(pred_path[ind])
    if len(top_n_pred) == n:
        break
```

0.279 為驗證任務的閾值

```python
%matplotlib inline 
print('0.279 為驗證任務的閾值, 即>= 0.279 可視為同一隻寵物')
print('上傳的相片:', src)

for c, p in zip(top_n_cos, top_n_pred):
    print(f'和 {p} 的相似度: {c}')

img_num = len(test_imgs)
for i, img in enumerate(test_imgs):
    plt.subplot(1 + img_num // 3, 3, i+1)
    plt.imshow(img[...,::-1])
    plt.title('upload img')
    plt.axis('off')
plt.show()

plt.figure(figsize=(11, 5))
for i in range(n):
    img = cv2.imread(top_n_pred_path[i])[...,::-1]
    plt.subplot(1, 5, i+1)
    plt.axis('off')
    plt.imshow(img)
    plt.title('%.3f' % top_n_cos[i])
```
